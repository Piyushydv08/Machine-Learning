```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('/content/train.csv')
df.head()
```

# 1. Categoriacl Data

```python
sns.countplot(df, x='Survived')
df['Survived'].value_counts()
# df['Survived'].value_counts().plot(kind='bar') ## upar vala ki tarah same chiz hai
```

```python
sns.countplot(df, x='Pclass')
```
