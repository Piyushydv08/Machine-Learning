# Connect to MySQL and query with Pandas

```python
import pandas as pd
from sqlalchemy import create_engine

# Create SQLAlchemy engine (root user with empty password)
conn = create_engine("mysql+pymysql://root:@localhost/world")
```

```python
df = pd.read_sql("SELECT * FROM city", conn)
df
```

```python
pd.read_sql_query("SELECT * FROM city WHERE CountryCode LIKE 'IND'", conn)
```
